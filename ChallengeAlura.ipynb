```python
# 1. Librerías necesarias (LangChain y el conector de Cohere)
!pip install -q langchain-cohere langchain

import os
from langchain_cohere import ChatCohere

# 2. API Key de Cohere
COHERE_API_KEY = "API_KEY_COHERE"
os.environ["COHERE_API_KEY"] = COHERE_API_KEY

# 3. Inicializar el modelo de Cohere
llm = ChatCohere(model="command-a-03-2025", temperature=0)

# 4. Función que le pregunta al agente
def preguntar_al_agente(pregunta_usuario):
    prompt = f"""
    Eres el Alura Agente experto en las políticas de la empresa NUTRIGENETICS.
    Responde a la siguiente pregunta utilizando ÚNICAMENTE la información del documento provisto abajo.
    Si la respuesta no está en el documento, di amablemente que no posees esa información.

    DOCUMENTO DE REFERENCIA:
    {contenido_procesado}

    PREGUNTA DEL USUARIO:
    {pregunta_usuario}

    RESPUESTA:
    """

    respuesta = llm.invoke(prompt)
    return respuesta.content

# 5. Prueba
pregunta = "¿Cuánto tiempo tengo para solicitar una devolución?"
print(f" Pregunta: {pregunta}")
print(f" Agente:\n{preguntar_al_agente(pregunta)}")
```
